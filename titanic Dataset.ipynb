# Logistic Regression

## one-hot-encoder 


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('Sex',OneHotEncoder(),[1]),
                  ('Embarked',OneHotEncoder(),[6])],
                 remainder='passthrough')
ip=np.array(ct.fit_transform(ip),dtype=str)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.15)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
```

```python
df=pd.DataFrame({'y':list(y_test),'prediction':yp})
```

```python
df
```

## Confusion Metrics

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1=ConfusionMatrixDisplay(cm)
cm1.plot()
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
precision=precision_score(y_test,yp)
f1=f1_score(y_test,yp)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy')

# train decision tree classifier
clf=clf.fit(x_train,y_train)

# predict the response of the test
y_pred=clf.predict(x_test)

from sklearn import metrics
accuracy2=metrics.accuracy_score(y_test,y_pred)
recall2=metrics=metrics.recall_score(y_test,y_pred,average='weighted')
print(recall2)
print(accuracy2)
```

# KNN Algorirhm

```python
from sklearn.neighbors import KNeighborsClassifier
alg2=KNeighborsClassifier(n_neighbors=9)
alg2.fit(x_train,y_train)
```

```python
yp2=alg2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc2=accuracy_score(y_test,yp2)
recall2=recall_score(y_test,yp2)
precision2=precision_score(y_test,yp2)
f12=f1_score(y_test,yp2)
```
